```
#Importando a tabela
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime, date, time
from sklearn.preprocessing import MinMaxScaler

df = spark.read.csv("/FileStore/tables/avocado.csv", header=True, sep=",")
df = df.toPandas()
df.head()
```

```
#Formatar colunas e tratar dados
df.rename(columns={'_c0': 'index'}, inplace = True)
df.head()
```

```
df2 = df.filter(['4046', '4225', '4770', 'year', 'Small Bags', 'Large Bags', 'XLarge Bags'])
df2["year"] = df2["year"].astype(int)
df2["4046"] = df2["4046"].astype(float)
df2["4225"] = df2["4225"].astype(float)
df2["4770"] = df2["4770"].astype(float)
df2["Small Bags"] = df2["Small Bags"].astype(float)
df2["Large Bags"] = df2["year"].astype(float)
df2["XLarge Bags"] = df2["XLarge Bags"].astype(float)

df2 = df2.groupby('year').agg(
    Pequeno=('4046','mean'),
    Medio=('4225','mean'),
    Grande=('4770','mean'),
    SmallBags=('Small Bags','mean'),
    LargeBags=('Large Bags','mean'),
    XLargeBags=('XLarge Bags','mean')
    ).reset_index()
df2
```

```
df2 = df2/1000
df2["year"] = df2["year"]*1000
df2 = df2.astype(int)

COLUNAS = ['Ano','Media','Classe']
new = pd.DataFrame(columns=COLUNAS)

def adicionar(classe, inicia):
    novaBase = df2.filter(['year', classe])
    novaBase["Classe"] = classe
    
    for i, infos in novaBase.iterrows():
        new.loc[inicia] = [infos[0], infos[1], infos[2]]
        inicia += 1
    
adicionar("Pequeno", 0)
adicionar("Medio", 4)
adicionar("Grande", 8)
adicionar("SmallBags", 12)
adicionar("LargeBags", 16)
adicionar("XLargeBags", 20)

new.head()
```
